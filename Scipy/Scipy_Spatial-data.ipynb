```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import Delaunay, ConvexHull
```

```python
arr= np.random.binomial(n=10, p=0.5, size=(10,2))
simplices= Delaunay(arr).simplices

plt.triplot(arr[:,0], arr[:,1], simplices)
plt.plot(arr[:,0], arr[:,1], color='red')
plt.show()
```

```python
hull = ConvexHull(arr)

for simplex in hull.simplices:
    plt.plot(arr[simplex, 0], arr[simplex, 1], 'k-')
plt.plot(arr[:,0], arr[:,1], 'o')
plt.show()
```
