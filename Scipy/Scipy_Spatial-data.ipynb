```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import Delaunay, ConvexHull, KDTree
```

```python
arr= np.random.binomial(n=10, p=0.5, size=(10,2))
simplices= Delaunay(arr).simplices

plt.triplot(arr[:,0], arr[:,1], simplices)
plt.plot(arr[:,0], arr[:,1], color='red')
plt.show()
```

```python
hull = ConvexHull(arr)

for simplex in hull.simplices:
    plt.plot(arr[simplex, 1], arr[simplex, 0], 'k-')
plt.plot(arr[:,1], arr[:,0], 'o')
plt.show()
```

```python
points= ([1,3],[2,5],[3,7],[4,9])

kdt= KDTree(points)
result= kdt.query([1,2])
print(result)

print("\nKDtree of points:", kdt)
```
